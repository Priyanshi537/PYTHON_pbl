# Phase 1: Data Cleaning and Preprocessing

This notebook performs the first phase of the project using Python only.

```python
from pathlib import Path
import numpy as np
import pandas as pd
```

```python
base_dir = Path.cwd().parent if Path.cwd().name == 'Phase_1' else Path.cwd()
dataset_path = base_dir / 'AirQualityUCI.csv'
phase_output = base_dir / 'Phase_1' / 'improved_dataset_phase_1.csv'

df = pd.read_csv(dataset_path, sep=';')
df = df.dropna(axis=1, how='all')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.columns = [column.strip() for column in df.columns]

for column in df.columns:
    if df[column].dtype == object:
        df[column] = df[column].astype(str).str.replace(',', '.', regex=False)

df['datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H.%M.%S', errors='coerce')
df = df.dropna(subset=['datetime']).sort_values('datetime').reset_index(drop=True)

numeric_columns = [column for column in df.columns if column not in {'Date', 'Time', 'datetime'}]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df[numeric_columns] = df[numeric_columns].replace(-200, np.nan)

df.to_csv(phase_output, index=False)
df.head()
```
